```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from labellines import labelLine, labelLines
import urllib.request
```

```python
urls = {
    2018: "https://raw.githubusercontent.com/Bogdanp/awesome-advent-of-code/master/2018.md",
    2019: "https://raw.githubusercontent.com/Bogdanp/awesome-advent-of-code/master/2019.md",
    2020: "https://raw.githubusercontent.com/Bogdanp/awesome-advent-of-code/master/2020.md",
    2021: "https://raw.githubusercontent.com/Bogdanp/awesome-advent-of-code/master/2021.md",
    2022: "https://raw.githubusercontent.com/Bogdanp/awesome-advent-of-code/master/README.md",
}
```

```python
langs = {}
for year in range(2018, 2023):
    url = urls[year]

    with urllib.request.urlopen(url) as response:
        data = response.read().decode("utf-8")

    lines = [i for i in data.splitlines() if i != ""]
    # print(lines)

    start = lines.index('### Solutions') + 1
    end = lines.index('### Live Streams')

    for line in lines[start:end]:
        if line.startswith("#### "):
            lang = line[5:].strip()
            if lang not in langs:
                langs[lang] = {}
                for y in range(2018, 2023):
                    langs[lang][y] = 0
        elif line.startswith("* ["):
            langs[lang][year] += 1

# print(langs)
```

```python
langs_items = langs.items()
langs_items = sorted(langs_items, key=lambda x: x[1][2022], reverse=True)

plt.figure(figsize=(16, 20))
plt.xticks(range(2018, 2023))
for lang, data in langs_items:
    plt.plot(data.keys(), data.values(), label=lang)
labelLines(plt.gca().get_lines(), zorder=2.5)
```
